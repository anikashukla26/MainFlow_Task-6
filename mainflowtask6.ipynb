```python
questions = ["1. What is the distribution of content types (movies vs TV shows) in the dataset?",
             "2. Which countries have produced the most Disney+ content?",
             "3. What are the most common ratings assigned to the content?",
             "4. What is the percentage distribution of content types (Movies vs TV Shows) on Disney+?",
             "5. What is the distribution of movie durations on Disney+?"]
questions
```

```python
content_type_distribution = df['type'].value_counts()

# Plotting the distribution using matplotlib
plt.figure(figsize=(8, 5))
content_type_distribution.plot(kind='bar', color=['lightblue', 'orange'])
plt.title('Distribution of Content Types (Movies vs TV Shows)', fontsize=16)
plt.xlabel('Content Type', fontsize=12)
plt.ylabel('Number of Titles', fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

```python
# Split the 'country' column (as some entries have multiple countries)
df['country'] = df['country'].fillna('Unknown')
country_data = df['country'].str.split(',', expand=True).stack().str.strip()

# Get the count of content produced by each country
country_distribution = country_data.value_counts().head(10)  # Top 10 countries

# Plotting the country distribution
plt.figure(figsize=(10, 6))
country_distribution.plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries Producing the Most Disney+ Content', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Number of Titles', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Fill missing values in the 'rating' column with 'Unknown'
df['rating'] = df['rating'].fillna('Unknown')

# Get the count of each rating
rating_distribution = df['rating'].value_counts()

# Plotting the rating distribution
plt.figure(figsize=(10, 6))
rating_distribution.plot(kind='bar', color='lightcoral')
plt.title('Most Common Ratings Assigned to Disney+ Content', fontsize=16)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Number of Titles', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Group by content 'type' and count the occurrences of each type
content_type_distribution = df['type'].value_counts()

# Plotting the distribution as a pie chart
plt.figure(figsize=(7, 7))
content_type_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'], explode=(0.05, 0.05))
plt.title('Percentage Distribution of Content Types (Movies vs TV Shows)', fontsize=16)
plt.ylabel('')  # Remove y-label for better presentation
plt.show()
```

```python
# Filter only the movies and extract durations
movies_data = df[df['type'] == 'Movie']

# Extract the duration in minutes from the 'duration' column and convert to integer
movies_data['duration'] = movies_data['duration'].str.replace(' min', '').astype(int)

# Plotting the distribution of movie durations using a histogram
plt.figure(figsize=(10, 6))
plt.hist(movies_data['duration'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Movie Durations on Disney+', fontsize=16)
plt.xlabel('Duration (Minutes)', fontsize=12)
plt.ylabel('Number of Movies', fontsize=12)
plt.grid(True)
plt.show()
```
